```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

df = pd.read_csv("CoursesData.csv")

# Replace NaN values in the "Description" column with empty strings
df["Description"] = df["Description"].fillna("")

# Convert course descriptions into numerical vectors
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df["Description"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_courses(user_input):
    user_input_vec = tfidf.transform([user_input])  # Convert input into numerical format
    similarities = cosine_similarity(user_input_vec, tfidf_matrix)  # Compute similarity
    recommended_indices = similarities.argsort()[0][-3:][::-1]  # Top 3 recommendations

    # Access columns using actual names from the DataFrame (CourseTitle)
    recommendations = df.iloc[recommended_indices][["CourseTitle", "Availability", "Cost"]]

    return recommendations

# Test in Jupyter or Google Colab
print(recommend_courses("I want to learn artificial intelligence"))
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Download the 'punkt_tab' resource
nltk.download("punkt_tab")

def chatbot_response(user_input):
    user_tokens = word_tokenize(user_input.lower())

    if "recommend" in user_tokens or "course" in user_tokens:
        recommendations = recommend_courses(user_input)
        return recommendations.to_string(index=False)
    elif "hello" in user_tokens or "hi" in user_tokens:
        return "Hello! I can recommend courses based on your interests. What would you like to learn?"
    else:
        return "Sorry, I didn't understand. Can you rephrase your query?"

# Test the chatbot in Jupyter
print(chatbot_response("Can you recommend a free AI course?"))
```

```python
import gradio as gr

def chat_interface(user_input):
    return chatbot_response(user_input)

iface = gr.Interface(fn=chat_interface, inputs="text", outputs="text")
iface.launch(share=True)  # Shareable link
```
